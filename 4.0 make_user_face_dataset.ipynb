```python
import numpy as np
import cv2
import os
import glob

cap = cv2.VideoCapture(0) # 노트북 웹캠을 카메라로 사용
cap.set(3,640) # 너비
cap.set(4,480) # 높이

user_name = "jinho" # 사진 찍는 사람 이름

if not os.path.isdir('User_image/'): # 폴더 없으면 폴더 생성
    os.mkdir('User_image/')

while(True):
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1) # 좌우 대칭

    cv2.imshow('frame', frame)
    
    k = cv2.waitKey(30) & 0xff
    if k == 27: # Esc 키를 누르면 종료
        break
    elif k == 49: # 1 키를 누르면 사진 찍음.
        if not os.path.isdir('User_image/{}/'.format(user_name)): 
            os.mkdir('User_image/{}/'.format(user_name)) # 폴더 없으면 폴더 생성
        
        i = len(glob.glob('User_image/{}/*'.format(user_name))) 
        cv2.imwrite('User_image/{}/{}.jpg'.format(user_name, i), frame)
    
cap.release()
cv2.destroyAllWindows()
```
